1\. seleccionar los productos cuyo nombre (Name) tenga cinco caracteres

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE LEN(NAME) = 5
```

2\. seleccionar los productos cuyo color sea Rojo (Red)

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE Color = 'red'
```

3\. seleccionar los productos cuyo color sea Azul (Blue) y ordenados por Name

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE Color = 'Blue' ORDER BY NAME
```

4\. seleccionar los productos cuyo color sea Azul o Rojo

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE Color = 'Blue' OR Color = 'Red'
```

5\. seleccionar los productos cuyo nombre (Name) tenga nueve caracteres y sean color 

plata (Silver)

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE LEN(Name) = 9 AND Color = 'SILVER'
```

6\. seleccionar los productos cuyo color sea Negro o Plata y tenga 11 caracteres.

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE LEN(Name) = 11 AND (Color = 'SILVER' OR Color = 'BLACK')
```

7\. seleccionar todos los productos cuya SellStartDate sea 2008-04-30

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE YEAR(SELLSTARTDATE) = 2008 AND MONTH(SELLSTARTDATE) = 04 AND  DAY(SELLSTARTDATE) = 30
```

14\. seleccionar todos los productos cuya SellstartDate sea del año 2008 y comiencen por 

C, terminan por g y son de color Plata (Silver

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE YEAR(SellStartDate) = 2008 AND Name LIKE 'C%G' AND COLOR = 'SILVER'
```

15\. seleccionar todos los productos y cuyo peso Weight esté entre 15 y 20.

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE Weight BETWEEN 15 and 20
```

16\. seleccionar todos los productos cuyo peso Weight esté entre 15 y 20 y su Size es 42,44 

o 46.

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE Weight BETWEEN 15 and 20 and Size = 42.44
```

17\. seleccionar todos los productos cuyo SellStartDate esté entre 2008-04-30 y una 

semana después

```sql
SELECT * FROM AdventureWorks2016.Production.Product as producto WHERE SellStartDate >= 28-04-30 and SellStartDate  <= DATEADD(WEEK, 1, '2008-04-30')
```
